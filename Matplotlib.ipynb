```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5,6]
y = [4,6,8,10,12,1]
```

```python
plt.xlabel('Days')
plt.ylabel('Stocks')
plt.plot(x,y, marker= 'o', linestyle='-', color='green', linewidth = 0.5)
plt.title('Sensex')
plt.show()
```

```python
plt.xlabel('Days')
plt.ylabel('Stocks')
plt.plot(x,y, marker= 'o', linestyle='-', color='green', linewidth = 0.5)
plt.title('Sensex')
plt.grid(axis = 'y')
plt.show()
```

Multi plots

```python
import numpy as np

x= np.array([1,7,4,5,9,3])
y=np.array([11,17,6,14,15,16,13,9])
z=np.array([1,2,3,5,6,7,10])

plt.plot(x,label="x")
plt.plot(y, label="y")
plt.plot(z, label="z")

plt.grid(axis='y')
plt.show()
```

Subplots

```python
import numpy as np
#plot1
x= np.array([1,7,4,5,9,3,12])
y=np.array([11,17,6,14,15,16,13])
# plt.subplot(r,c,pos)
plt.subplot(1,2,1)
plt.plot(x,y)
#plot2
y=np.array([11,17,6,14,15,16,13])
z=np.array([1,2,3,5,6,7,10])
plt.subplot(1,2,2)
plt.plot(y,z)

plt.show()
```

```python
import numpy as np
x= np.array([1,2,3,4,5,6,7])
y=x
z= np.array([10,300,40,60,100,1000,120])
colors =['green']* (len(x)-1)
colors.append('red')
plt.scatter(x,y, c= colors, s=z, alpha =0.45)

plt.show()
```
